1) Как бы Вы организовали аналитику по данному продукту?

Мне нравится подход CRISP DM 

Бизнес-анализ 

 Всегда важно помнить, что "профессию Data Scientist" - это продукт, а на продукт следует смотреть не перспективой продукта как такового, а глазами потребителя. 
 
  Какую нужду/желание данный продукт удовлетворяет?
   " Нетология продает продукт "профессия Data Scientist" который по своим параметрам не уступает тому, что предлагают конкуренты" - не правильный подход. Who cares?
   
     1.1 - Определение бизнес-целей.
     Цель- дать возможность нашему потребителю получить современную востребованную специальность, которая даст ему/ей конкурентное преимущество на рыкне вакансий. 
     1.2 - Оценка текущей ситуации.
     Пока что ситуация для данного продукта благоприятная, спрос опережает предложение. Но это нормальная ситуация описанная в бизнес литературе, т.н. модель S .
     1.3 - Клиент приходит часто с нулевой базой, на выходе должны получить "desk-ready" специалиста. Если наш продукт не достигнет поставленных результатов, получим неудовлетворенного клиента и конкуренты сметут нас с рынка. 
     Значит требуется определить ключевые компоненты данного продукта, т.е. то, что сделает, в сумме, клиента готовым специалистом на рынке вакансий.
     
2) Сформулируйте 2-3 ключевые метрики продукта

    -  На какой жизненный цикл расчитан продукт? Определить где какие стадии данного цикла находятся. Допустим сейчас стадия активного роста. Когда начнется насыщенность? Когда может наступить стадия старения и упадка. Как данный продукт может эволюционировать?
    
    - Какую долю рынка мы занимаем и в чем наше конкурентное преимущество? Если у нас продукт не хуже, чем у конкурентов, значит нет ни какого преимущества, просто мы на волне роста и нам повезло. Успех надо закреплять долгосрочным планированием и бренд менеджментом, для этого необходима аналитика самого продукта, точнее компонентов, также аналитика клиентов и их причин потребления продукта.
    
     - "Кастомизация" продукта. Насколько наш продукт гибок к каждому клиенту? Или у нас стандартный эконом-продукт уровня "учи или свали". Аналитика должна выявить такой требуемый набор компонентов продукта, чтобы суммарно клиент получал синергию потребителя, которая и выльется в конкурентное преимущество.
